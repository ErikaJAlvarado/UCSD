```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=["day"])

#df["day"] = pd.to_datetime(df["day"])
type(df["day"][0])
df.set_index("day",inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({"temperature":0,"windspeed":0,"event":"no event"})
new_df
```

```python
#forward fill
new_df = df.fillna(method="ffill")
new_df
```

```python
#Backward fill

new_df = df.fillna(method="bfill")
new_df
```

```python
#Backwards at the column level
new_df = df.fillna(method="bfill", axis = "columns")
new_df
```

```python
#limit copies the data only a number of time.  In the case below it copied only once.
new_df = df.fillna(method = "ffill", limit =1)
new_df
```

```python
#linear interpolation
new_df = df.interpolate()
new_df
```

```python
#Interpolate values taking in consideration how far or close the missing data is next to the next date.
new_df = df.interpolate(method = "time")
new_df
```

```python
#Dropping na values.  It drop all rows that have at least one na
new_df = df.dropna()
new_df
```

```python
#Dropping only rows that has all columns na
new_df = df.dropna(how = "all")
new_df
```

```python
#dropna with threshold.  Don't drop rows that have at least "n"  na values.  rows with n+ na values are dropped.
new_df = df.dropna(thresh=1)
new_df
```

```python
#dropna with threshold.  Don't drop rows that have at least "n"  na values.  rows with n+ na values are dropped.
new_df = df.dropna(thresh=2)
new_df
```

```python
#Insert missing dates
dt = pd.date_range("01-01-2017","01-11-2017")
type(dt)
idx = pd.DatetimeIndex(dt)
type(idx)

new_df = df.reindex(idx)
new_df
```
